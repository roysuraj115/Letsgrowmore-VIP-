```python
#importing some basic libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#loading the data
df=pd.read_csv("global.csv",encoding="latin1")
df.head()
```

```python
#perfoming some basic operation on the dataset
df.info()
```

```python
#checking for the null value
df.isnull().sum()
```

```python
df.columns
```

```python
#copying the dataframe and renaming the columns
df1=df.copy()
df1.rename(columns={'iyear':'Year','imonth':'Month','iday':'Day','country_txt':'Country','provstate':'state',
                       'region_txt':'Region','attacktype1_txt':'AttackType','target1':'Target','nkill':'Killed',
                       'nwound':'Wounded','summary':'Summary','gname':'Group','targtype1_txt':'Target_type',
                       'weaptype1_txt':'Weapon_type','motive':'Motive'},inplace=True)
```

#plotting graphs to better understand the data


```python
#plotting Country with the most attacks
plt.figure(figsize=(22,10))
sns.barplot(x=df1.Country.value_counts().head(10).index,y=df1.Country.value_counts().head(10).values)
plt.title("Country with the most attacks")
plt.xlabel("countries")
plt.ylabel("No. of terrorist activities")
```

```python
#plotting City with the most attacks
plt.figure(figsize=(22,10))
sns.barplot(x=df1.state.value_counts().head(10).index,y=df1.state.value_counts().head(10).values)
plt.title("City with the most attacks")
plt.xlabel("Cities")
plt.ylabel("No. of Attacks/terrorist activities")
```

```python
#plotting Region with the most attacks
plt.figure(figsize=(22,10))
sns.barplot(x=df1.Region.value_counts().head(10).index,y=df1.Region.value_counts().head(10).values)
plt.title("Region with the most attacks")
plt.xlabel("Regions")
plt.ylabel("No. of Attacks/terrorist activities")
```

```python
#plotting Year with the most attacks
plt.figure(figsize=(22,10))
sns.barplot(x=df1.Year.value_counts().index,y=df1.Year.value_counts().values)
plt.title("Year with the most attacks")
plt.xlabel("Year")
plt.ylabel("No. of Attacks/terrorist activities")
```

```python
#plotting Months with the most attacks
plt.figure(figsize=(20,6))
sns.barplot(x=df1.Month.value_counts().index,y=df1.Month.value_counts().values)
plt.title("Month with the most attacks")
plt.xlabel("Month")
plt.ylabel("No. of Attacks/terrorist activities")
```

```python
#plotting Group with most attacks
plt.figure(figsize=(25,10))
sns.barplot(x=df1.Month.value_counts().head(10)[1:].values,y=df1.Group.value_counts().head(10)[1:].index)
plt.title("Group with the most attacks")
plt.xlabel("No. of Attacks/terrorist activities")
plt.ylabel("Group")
```

```python
#plotting Most Attack Types
plt.figure(figsize=(22,10))
sns.barplot(x=df1.AttackType.value_counts().index,y=df1.AttackType.value_counts().values)
plt.title("Most type of attacks")
plt.xlabel("Attack types")
plt.ylabel("Count")
```

```python
#plotting Most Target Types
plt.figure(figsize=(22,10))
sns.barplot(x=df1.Target_type.value_counts().values,y=df1.Target_type.value_counts().index)
plt.title(" Target with the most attacks")
plt.xlabel("count")
plt.ylabel("Target_type")
```

```python
#printing insights
print("Country with the most attacks:",df1['Country'].value_counts().idxmax())
print("City with the most attacks:",df1['city'].value_counts().index[1]) #as first entry is 'unknown'
print("Region with the most attacks:",df1['Region'].value_counts().idxmax())
print("Year with the most attacks:",df1['Year'].value_counts().idxmax())
print("Month with the most attacks:",df1['Month'].value_counts().idxmax())
print("Group with the most attacks:",df1['Group'].value_counts().index[1])
print("Most Attack Types:",df1['AttackType'].value_counts().idxmax())
```
